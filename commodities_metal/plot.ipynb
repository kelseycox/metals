```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import datetime 
```

```python
# Load in CSVs
gold = pd.read_csv("csvs/gold.csv")
gold
```

```python
# Drop unwanted columns
drop_gold= gold.drop(columns={'Unnamed: 0'})
drop_gold
```

```python
# Rename price as metal
new_gold= drop_gold.rename(columns={'Price (USD)' : "Gold"})
new_gold
```

```python
silver= pd.read_csv("csvs/silver.csv")
drop_silver= silver.drop(columns={'Unnamed: 0', 'Date'})
drop_silver
```

```python
# Rename price as metal
new_silver= drop_silver.rename(columns={'Price (USD)' : "Silver"})
new_silver
```

```python
palladium= pd.read_csv("csvs/palladium.csv")
palladium
```

```python
# Drop unwanted columns
drop_palladium= palladium.drop(columns={'Unnamed: 0', 'Date'})
drop_palladium
```

```python
# Rename price as metal
new_palladium= drop_palladium.rename(columns={'Price (USD)': 'Palladium'})
new_palladium
```

```python
platinum=pd.read_csv("csvs/platinum.csv")
platinum
```

```python
# Drop unwanted columns
drop_platinum= platinum.drop(columns={'Unnamed: 0' , 'Date'})
drop_platinum
```

```python
# Rename price as metal
new_platinum= drop_platinum.rename(columns={'Price (USD)': 'Platinum'})
new_platinum
```

```python
# Merge all DataFrames
merged_df = pd.concat([new_gold,new_silver,new_palladium,new_platinum], axis=1)
merged_df
```

```python
final_df= merged_df.dropna()
final_df
```

```python
# Export to CSV
final= final_df.to_csv('final.csv')
final
```

```python
# Plot

ax1 = final_df.plot(y='Gold', kind="line", color="yellow", label="Gold", title="Metals by Price (USD)")
final_df.plot(y='Silver', kind="line", color="gray", label="Silver", ax=ax1)
final_df.plot(y='Palladium', kind="line", color="cyan", label="Palladium", ax=ax1)
final_df.plot(y='Platinum', kind="line", color="pink", label="Platinum", ax=ax1)

# Plot labels
ax1.set_xlabel("Date")
ax1.set_ylabel("Price in USD")

# Plot the legend
ax1.legend(loc="best")

# Display graph
plt.savefig("Metals.png")
plt.show()
```
